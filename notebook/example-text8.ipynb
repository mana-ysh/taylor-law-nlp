```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import sys

sys.path.append("../src/")
from taylor import TaylorLowRegressor
```

```python
DATA_PATH = "../data/text8"
dat = open(DATA_PATH).read().split()
print("length of text: {}".format(len(dat)))
```

```python
m = TaylorLowRegressor()
m.fit(dat, 5620)
```

```python
print("c: {}".format(m.res[0][0]))
print("sigma: {}".format(m.res[0][1]))
```

```python
df = pd.DataFrame(m.stats)
df
```

```python
plt.title("Taylor's Law Example with text8 corpus")
plt.xlabel("$\mu$")
plt.ylabel("$\sigma$")
plt.xscale("log")
plt.yscale("log")
plt.plot(df.mu, df.sigma, 'o')
plt.plot(df.mu, m.res[0][0] * df.mu ** m.res[0][1])
# plt.plot(np.log10(df.mu**m.res[0][1]), np.log10(df.sigma), 'o')
```
